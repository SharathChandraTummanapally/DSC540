```python
import pandas as pd
import numpy as np
```

```python
# Load the data
candy_df = pd.read_excel('Data/candyhierarchy2017.xlsx')

# Validate the dataload
candy_df.head()
```

Fill in missing data

```python
# Getting missing data count.
candy_df.fillna(0)
```

Remove duplicates

```python
candy_df.drop_duplicates()
```

Chapter 8
Create hierarchial index

```python
# Creating the hierarchial index.
candy_h_df = candy_df.set_index(['Q1: GOING OUT?', 'Q2: GENDER', 'Internal ID'])

# Checking the dataframe.
candy_h_df.head()
```

Combine and Merge Datasets

```python
# Creating a database for demographics.
demo_candy_df = candy_df.iloc[:,:8]

# Removing Q1.
demo_candy_df = demo_candy_df.drop(['Q1: GOING OUT?'], axis=1)

# Checking the dataframe.
demo_candy_df.head()
```

```python
# Creating database for responses.
candy_resp_df = candy_df.drop(['Q2: GENDER', 'Q3: AGE', 'Q4: COUNTRY', 'Q5: STATE, PROVINCE, COUNTY, ETC'], axis=1)

# Checking the dataframe.
candy_resp_df
```

```python
# Merging the dataframes.
pd.merge(demo_candy_df, candy_resp_df, on='Internal ID')
```

Chapter 10
Grouping with Index Levels

```python
# Getting the responses by gender count.
candy_h_df.groupby(level='Q2: GENDER', axis=0).count()
```

Split/Apply/Combine

```python
# Function to obtain the 5 youngest records.
def oldest(df, n=5, column='Q3: AGE'):
    return df.sort_values(by=column)[:n]

# Replace any strings in the Age column with NAN
candy_df['Q3: AGE'] = candy_df['Q3: AGE'].apply(pd.to_numeric,errors='coerce')

# Apply the function after grouping by gender
candy_df.groupby('Q2: GENDER').apply(oldest)
```

Chapter 11

```python
# Loading the datafile.
boing_candy_df = pd.read_excel(r'Data/BOING-BOING-CANDY-HIERARCHY-2016-SURVEY-Responses.xlsx')

# Checking the dataframe.
boing_candy_df.head()
```

Convert between string and date time

```python
# Converting to string.
boing_candy_df['string time'] = boing_candy_df['Timestamp'].astype(str)

# View the new column.
boing_candy_df['string time']
```

```python
# Converting string to date time.
boing_candy_df['Time'] = pd.to_datetime(boing_candy_df['string time'])

# View the new column.
boing_candy_df['Time']
```

Convert timestamps to periods and back

```python
# Setting Timestamp as the index.
boing_candy_df = boing_candy_df.set_index(['Timestamp'])

# Converting to periods with a daily frequency.
period_df = boing_candy_df.to_period(freq='D')
# view the dataframe
period_df
```

```python
# Convert from daily period to timestamp at end of day.
ts_df = period_df.to_timestamp(how='end')

# Checking the dataframe.
ts_df.head()
```
